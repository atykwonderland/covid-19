```python
import pandas as pd

# Import Data
df = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
df_world = df.copy()

# Get Input
province = input("province/state (if none, write NA): ") 
country = input("country/region: ") 
date = input("date (write as mm/dd/yyyy): ") 

#Clean data to match inputs
df_world = df_world.drop(['SNo'], axis=1)
df_date = df_world[df_world['ObservationDate'] == date]
df_country_date = df_date[df_date['Country/Region'] == country]

#Get sums of Confirmed, Deaths, Recovered
if province == "NA":
    #add up add regions in country for cases
    total = df_country_date.sum(axis = 0, skipna = True, numeric_only = True)
else:
    df_province_date = df_country_date[df_country_date['Province/State'] == province]
    total = df_province_date.sum(axis = 0, skipna = True, numeric_only = True)

#Print the totals
print(total)
```
